```python
class employee:
    def __init__(self):
        self.emp_id=[]
        self.name=[]
        self.gender=[]
        self.age=[]
        self.city=[]
        self.salary=[]
        self.total_exp=[]
        self.emp_for_comp=[]
        
    def set_data(self):
        count=0
        while count<5:#taking 5 to save time            
            self.emp_id.append(int(input("Id number: ")))#int only
            self.name.append(input("Name: "))
            self.gender.append(input("Gender: "))
            self.age.append(int(input("Age: ")))
            self.city.append(input("City: "))
            self.salary.append(int(input("Salary : ")))
            self.total_exp.append(int(input("Total Experience")))
            self.emp_for_comp.append(int(input("Employee for company from how many years")))
            count+=1
        data={"id":self.emp_id,
              "name":self.name,
              "gen":self.gender,
              "age":self.age,
              "city":self.city,
              "sal":self.salary,
              "total_exp":self.total_exp,
              "emp_comp":self.emp_for_comp
             }
        
        import pandas as pd
        db=pd.DataFrame(data)
        db.to_csv("employee_data_25.csv")
        print(db)
        return db
        
        
        
        


```
